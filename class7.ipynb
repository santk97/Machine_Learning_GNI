```python
survived=df[df['Survived']==1]
not_Survived=df[df['Survived']==0]
print "survived details ", survived.head()
print '\n\n not survived details ', not_Survived.head()
```

```python
#print the details 

print 'Survived : %i %.2f%%'  %(len(survived) , float(len(survived))/len(df)*100)
print ' not Survived : %i %.2f%%'  %(len(not_Survived) , float(len(not_Survived))/len(df)*100)
```

```python
#survival chances based on PAssenger class
df.Pclass.value_counts()
df.groupby('Pclass').Survived.value_counts()
```

```python
#survival chances based on Sex of the passenger
df.Sex.value_counts()
df.groupby('Sex').Survived.value_counts()
```

```python
#survival chances of passengers based on their age
df.Age.value_counts()
df.groupby('Age').Survived.value_counts()
```

```python
#survival chances of passenger based on Fare
df.Fare.value_counts()
df.groupby('Fare').Survived.value_counts()
```

```python
#survival chances based on the Emabrked
df.Embarked.value_counts()
df.groupby('Embarked').Survived.value_counts()
```
